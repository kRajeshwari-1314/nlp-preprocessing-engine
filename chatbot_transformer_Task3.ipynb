## Step 2: Import Libraries

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
```

###Step 3: Load Pre-trained Transformer Model and Tokenizer

```python
model_name = "Qwen/Qwen2.5-0.5B-Instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype="auto",
    device_map="auto"
)

print("Model loaded successfully!")
```

## Step 4: Creating Response Logic

In this step, we define how the chatbot processes user input and produces output.

• Formats the conversation into a structure the model understands  
• Sends the processed input to the transformer model  
• Generates a response based on context  
• Decodes the output into readable text  

```python
def get_response(conversation):
    formatted_text = tokenizer.apply_chat_template(
        conversation,
        tokenize=False,
        add_generation_prompt=True
    )

    inputs = tokenizer([formatted_text], return_tensors="pt").to(model.device)

    outputs = model.generate(
        **inputs,
        max_new_tokens=80,
        do_sample=True,
        temperature=0.7,
        top_p=0.9
    )

    outputs = [
        out_ids[len(in_ids):]
        for in_ids, out_ids in zip(inputs.input_ids, outputs)
    ]

    reply = tokenizer.batch_decode(outputs, skip_special_tokens=True)[0]

    return reply
```

## Step 5: Chatbot Interaction Loop
The chatbot continuously interacts with the user until the user types **exit** or **quit**.

```python
print("Chatbot: Hi! I am your assistant. Type 'exit' or 'quit' to end the chat.")

conversation = [
    {"role": "system", "content": "You are a helpful AI assistant."}
]

while True:
    user_input = input("You: ")

    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot: Session ended. Goodbye!")
        break

    conversation.append({"role": "user", "content": user_input})

    reply = get_response(conversation)

    print("Chatbot:", reply)

    conversation.append({"role": "assistant", "content": reply})
```

## Step 6: Pipeline Flow Explanation
User Input → Tokenization → Model Processing → Response Generation → Output Display → Repeat Until Exit

##  How It Works

• The user provides input through the console  
• The text is converted into tokens using the tokenizer  
• The transformer model processes the tokens along with conversation history  
• A response is generated based on learned patterns  
• The output is converted back into readable text  
• The chatbot continues the conversation until the user exits
