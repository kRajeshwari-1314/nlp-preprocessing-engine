## Step 2: Import Required Libraries

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
```

###Step 3: Load Pre-trained Transformer Model and Tokenizer

```python
model_name = "Qwen/Qwen2.5-0.5B-Instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype="auto",
    device_map="auto"
)

print("Model loaded successfully!")
```

###Step 4: Define Chatbot Response Function

This function:
1. Takes user input  
2. Adds it to conversation history  
3. Generates a response using the transformer model  
4. Returns the chatbot reply along with updated conversation history

```python
def generate_response(messages):
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )

    model_inputs = tokenizer([text], return_tensors="pt").to(model.device)

    generated_ids = model.generate(
        **model_inputs,
        max_new_tokens=80,
        do_sample=True,
        temperature=0.7,
        top_p=0.9
    )

    generated_ids = [
        output_ids[len(input_ids):]
        for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
    ]

    response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]

    return response
```

###Step 5: Build Interactive Chatbot Loop

The chatbot keeps running until the user types exit or quit.

```python
print("Chatbot: Hello! I am your AI assistant. Type 'exit' or 'quit' to stop.")

messages = [
    {"role": "system", "content": "You are a helpful and friendly AI assistant."}
]

while True:
    user_input = input("You: ")

    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot: Goodbye! Have a great day.")
        break

    messages.append({"role": "user", "content": user_input})

    response = generate_response(messages)

    print("Chatbot:", response)

    messages.append({"role": "assistant", "content": response})
```

## Step 6: Explanation of the Workflow:

Pipeline Flow
**User Input → Tokenization → Model Processing → Response Generation → Display Output → Repeat until Exit**

How it Works
- The user enters a message.
- The tokenizer converts text into token IDs.
- The model processes the input and generates a response.
- Conversation history is stored so the chatbot can continue the discussion naturally.
- The loop runs until the user enters exit or quit.
